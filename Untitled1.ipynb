```python
try:
    # Gunakan path lengkap untuk memastikan file ditemukan
    df = pd.read_csv('/shopee_ulasan.csv')
    print("✅ Dataset berhasil dimuat.")
except FileNotFoundError:
    print("❌ File 'shopee_ulasan.csv' tidak ditemukan. Pastikan Anda telah mengunggahnya ke Google Colab.")
    df = pd.DataFrame({'ulasan': [], 'rating': []}) # Buat dataframe kosong

# Tampilkan 5 baris pertama data
print("\nLima baris pertama data (head):")
print(df.head())

# Tampilkan informasi dasar tentang dataset
print("\nInformasi dasar dataset (info):")
df.info()
```

```python
print("\nMemulai proses pembersihan data...")

# Cek nilai yang hilang (missing values)
df.dropna(subset=['ulasan'], inplace=True)

# Cek dan hapus data duplikat
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

print(f"✅ Data bersih dan siap dianalisis. Jumlah data: {len(df)} baris.")
```

```python
print("\nMemulai analisis dengan AI menggunakan Replicate...")

# 4.1. Fungsi untuk Klasifikasi Teks
def classify_text(review_text):
    """
    Mengklasifikasikan teks ulasan menggunakan model via Replicate.
    """
    prompt = f"""Instruction: Classify the following Indonesian customer review as 'Positive', 'Negative', or 'Neutral'. Provide only the label.
Input: "{review_text}"
Output:"""

    try:
        # Menjalankan model di Replicate
        output = replicate.run(
            MODEL_ID,
            input={"prompt": prompt, "max_new_tokens": 10}
        )

        # Output dari replicate.run adalah sebuah iterator, kita gabungkan hasilnya
        result = "".join(output).strip().lower()

        if 'positive' in result or 'positif' in result: return 'Positive'
        if 'negative' in result or 'negatif' in result: return 'Negative'
        if 'neutral' in result or 'netral' in result: return 'Neutral'
        return 'Unclassified'

    except Exception as e:
        print(f"Terjadi error saat klasifikasi: {e}")
        return "Error"

# 4.2. Fungsi untuk Meringkas Teks
def summarize_text(review_text):
    """
    Meringkas keluhan utama dari teks ulasan negatif via Replicate.
    """
    prompt = f"""Instruction: Summarize the main complaint from the following Indonesian customer review in one short sentence.
Input: "{review_text}"
Output:"""

    try:
        # Menjalankan model di Replicate
        output = replicate.run(
            MODEL_ID,
            input={"prompt": prompt, "max_new_tokens": 50}
        )

        # Menggabungkan hasil dari iterator
        result = "".join(output).strip()
        return result

    except Exception as e:
        print(f"Terjadi error saat meringkas: {e}")
        return "Error"

# 4.3. Aplikasi pada Dataset
# Kita tetap gunakan sampel kecil (5 baris) untuk uji coba agar cepat.
df_sample = df.head(5).copy()

print("\nMemproses klasifikasi sentimen untuk 5 sampel data...")
df_sample['sentiment_category'] = df_sample['ulasan'].apply(classify_text)
print("Klasifikasi selesai.")

print("\nMemproses ringkasan untuk ulasan negatif pada sampel...")
if 'sentiment_category' in df_sample.columns:
    negative_reviews_mask = df_sample['sentiment_category'] == 'Negative'
    df_sample.loc[negative_reviews_mask, 'summary'] = df_sample.loc[negative_reviews_mask, 'ulasan'].apply(summarize_text)
else:
    print("Kolom 'sentiment_category' tidak ditemukan, ringkasan dilewati.")
print("Ringkasan selesai.")

# Tampilkan hasil pada sampel
print("\nHasil analisis pada sampel data:")
print(df_sample[['ulasan', 'sentiment_category', 'summary']].fillna(''))
```

```python
print("\nMembuat visualisasi dari hasil analisis...")

sentiment_counts = df_sample['sentiment_category'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="viridis")
plt.title('Distribusi Sentimen Ulasan Pelanggan Shopee', fontsize=16)
plt.xlabel('Kategori Sentimen', fontsize=12)
plt.ylabel('Jumlah Ulasan', fontsize=12)
plt.show()

summaries_text = ' '.join(df_sample[df_sample['sentiment_category'] == 'Negative']['summary'].dropna())
if summaries_text:
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(summaries_text)
    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud dari Keluhan Utama Pelanggan', fontsize=16)
    plt.show()
else:
    print("\nTidak ada ringkasan keluhan untuk dibuatkan Word Cloud.")
```
